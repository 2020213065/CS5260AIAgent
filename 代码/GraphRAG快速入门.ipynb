1. жһ¶жһ„еҚҮзә§дёҺзҹҘиҜҶеӣҫи°ұжһ„е»әдјҳеҢ–

- **жҷәиғҪеҢ–зҡ„зҹҘиҜҶеӣҫи°ұжһ„е»ә**  
  GraphRAG 2.0 еңЁж•°жҚ®йў„еӨ„зҗҶйҳ¶ж®өпјҢйҖҡиҝҮеӨ§жЁЎеһӢиҮӘеҠЁжҠҪеҸ–ж–Үжң¬дёӯзҡ„е®һдҪ“еҸҠе…ізі»пјҢжһ„е»әеҮәеұӮж¬ЎеҢ–зҡ„зҹҘиҜҶеӣҫи°ұгҖӮзӣёжҜ”дј з»ҹзҡ„з®ҖеҚ•ж–Үжң¬зүҮж®өжЈҖзҙўпјҢжһ„е»әеҗҺзҡ„еӣҫи°ұиғҪеӨҹд»ҘвҖңзӨҫеҢәвҖқпјҲtopic-based clustersпјүзҡ„ж–№ејҸеҜ№ж•°жҚ®иҝӣиЎҢз»„з»ҮпјҢиҝҷж ·дёҚд»…еҸҜд»ҘиҰҶзӣ–е…ЁеұҖдҝЎжҒҜпјҢд№ҹиғҪй’ҲеҜ№еұҖйғЁжҹҘиҜўз»ҷеҮәжӣҙзІҫеҮҶзҡ„зӯ”жЎҲгҖӮ  

- **еҠЁжҖҒзӨҫеҢәйҖүжӢ©жңәеҲ¶**  
  ж–°зүҲжң¬еј•е…ҘдәҶеҠЁжҖҒзӨҫеҢәйҖүжӢ©жөҒзЁӢгҖӮзі»з»ҹдјҡеңЁз”ҹжҲҗе“Қеә”д№ӢеүҚпјҢеҜ№зҹҘиҜҶеӣҫи°ұдёӯдёҚеҗҢвҖңзӨҫеҢәвҖқзҡ„зӣёе…іжҖ§иҝӣиЎҢиҜ„дј°пјҢд»ҺиҖҢд»…дҝқз•ҷдёҺеҪ“еүҚжҹҘиҜўжңҖеҢ№й…Қзҡ„йғЁеҲҶгҖӮиҝҷз§ҚжңәеҲ¶иғҪжңүж•ҲвҖңдёўејғвҖқеҷӘеЈ°ж•°жҚ®пјҢжҸҗй«ҳжЈҖзҙўж•ҲзҺҮе’Ңзӯ”жЎҲзҡ„еҮҶзЎ®жҖ§гҖӮ

2. жҹҘиҜўжөҒзЁӢдёҺжҲҗжң¬дјҳеҢ–

- **дёӨйҳ¶ж®өжҹҘиҜўжөҒзЁӢ**  
  GraphRAG 2.0 е°Ҷж•ҙдёӘжөҒзЁӢжӢҶеҲҶдёәвҖңзҙўеј•йҳ¶ж®өвҖқе’ҢвҖңжҹҘиҜўйҳ¶ж®өвҖқпјҡ  
  - еңЁзҙўеј•йҳ¶ж®өпјҢзі»з»ҹеҲ©з”ЁеӨ§жЁЎеһӢеҜ№еҺҹе§Ӣж•°жҚ®иҝӣиЎҢз»“жһ„еҢ–еӨ„зҗҶпјҢжҸҗеҸ–е®һдҪ“еҸҠе…¶е…ізі»пјҢжһ„е»әеҲҶеұӮзҹҘиҜҶеӣҫи°ұпјӣ  
  - еңЁжҹҘиҜўйҳ¶ж®өпјҢеҲҷе…ҲиҝӣиЎҢеҲқжӯҘзҡ„зӣёе…іжҖ§жөӢиҜ•пјҢеҶҚеҲ©з”Ёз»ҸиҝҮеҠЁжҖҒзӨҫеҢәзӯӣйҖүзҡ„дҝЎжҒҜжқҘз”ҹжҲҗдёҠдёӢж–Үдё°еҜҢгҖҒзІҫеҮҶзҡ„еӣһзӯ”гҖӮ  
  иҝҷз§ҚеҲҶжӯҘеӨ„зҗҶдёҚд»…жҸҗй«ҳдәҶжЈҖзҙўзҡ„е№ҝеәҰе’Ңж·ұеәҰпјҢиҝҳиғҪж №жҚ®жҹҘиҜўйңҖжұӮзҒөжҙ»и°ғз”ЁеӨ§жЁЎеһӢгҖӮ  

- **Tokenж¶ҲиҖ—еӨ§е№…йҷҚдҪҺ**  
  дёәдәҶеә”еҜ№еӨ§и§„жЁЎж•°жҚ®и°ғз”Ёж—¶й«ҳжҳӮзҡ„жҲҗжң¬пјҢGraphRAG 2.0 еҜ№ LLM зҡ„и°ғз”ЁеҒҡдәҶдјҳеҢ–гҖӮжҚ®жҠҘйҒ“пјҢеңЁжҹҗдәӣеңәжҷҜдёӢж•ҙдҪ“ Token ж¶ҲиҖ—йҷҚдҪҺй«ҳиҫҫ 77%пјҢиҝҷдҪҝеҫ—зі»з»ҹеңЁдҝқиҜҒй«ҳиҙЁйҮҸеӣһзӯ”зҡ„еҗҢж—¶пјҢд№ҹеӨ§е№…жҸҗеҚҮдәҶжҲҗжң¬ж•ҲзҺҮгҖӮ  

- **LazyGraphRAG жЁЎејҸ**  
  ж–°зүҲжң¬иҝҳжҺЁеҮәдәҶвҖңLazyGraphRAGвҖқжЁЎејҸвҖ”вҖ”дёҖз§Қз»“еҗҲдәҶеҗ‘йҮҸжЈҖзҙўе’Ңеӣҫз»“жһ„жЈҖзҙўдјҳеҠҝзҡ„ж–№жЎҲгҖӮиҜҘжЁЎејҸйҮҮз”Ёиҝӯд»Јж·ұеҢ–зҡ„ж–№ејҸпјҢеҸӘжңүеңЁеҝ…иҰҒж—¶жүҚи°ғз”Ёиө„жәҗеҜҶйӣҶеһӢзҡ„еӨ§жЁЎеһӢиҝӣиЎҢж·ұеәҰеҲҶжһҗпјҢд»ҺиҖҢе®һзҺ°дәҶдёҺдј з»ҹ GraphRAG зӣёжҜ”жҲҗжң¬жӣҙдҪҺдҪҶж•ҲжһңзӣёеҪ“зҡ„зӣ®ж ҮгҖӮ

#### 2.еҖҹеҠ©APIиҝӣиЎҢQueryиҝҮзЁӢ

жҹҘиҜўзҙўеј•

еңЁжҹҘиҜўзҙўеј•д№ӢеүҚпјҢеҝ…йЎ»е…Ҳе°ҶеӨҡдёӘзҙўеј•ж–Үд»¶еҠ иҪҪеҲ°еҶ…еӯҳдёӯпјҢе№¶дј йҖ’з»ҷжҹҘиҜў APIгҖӮ

````python
entities = pd.read_parquet(f"{PROJECT_DIRECTORY}/output/entities.parquet")
communities = pd.read_parquet(f"{PROJECT_DIRECTORY}/output/communities.parquet")
community_reports = pd.read_parquet(
    f"{PROJECT_DIRECTORY}/output/community_reports.parquet"
)
````

дёҠиҝ°д»Јз ҒиҜ»еҸ– .parquet ж јејҸзҡ„зҙўеј•ж•°жҚ®пјҢеҢ…жӢ¬ е®һдҪ“пјҲentitiesпјүгҖҒзӨҫзҫӨпјҲcommunitiesпјү д»ҘеҸҠ зӨҫзҫӨжҠҘе‘ҠпјҲcommunity_reportsпјүгҖӮ

жү§иЎҢе…ЁеұҖжҗңзҙў

````python
query = "иҜ·её®жҲ‘еҜ№жҜ”дёӢID3е’ҢC4.5еҶізӯ–ж ‘з®—жі•дјҳеҠЈеҠҝгҖӮе№¶з”Ёдёӯж–ҮиҝӣиЎҢеӣһзӯ”гҖӮ"
````

````python
response, context = await api.global_search(
    config=graphrag_config,
    entities=entities,
    communities=communities,
    community_reports=community_reports,
    community_level=2,
    dynamic_community_selection=False,
    response_type="Multiple Paragraphs",
    query=query,
)
````

еңЁиҝҷйҮҢпјҢжҲ‘д»¬и°ғз”Ё `global_search` ж–№жі•пјҢдҪҝз”Ёе·ІеҠ иҪҪзҡ„зҙўеј•ж•°жҚ®иҝӣиЎҢжҹҘиҜўгҖӮ  
- `community_level=2`пјҡи®ҫе®ҡзӨҫзҫӨеұӮзә§дёә 2 зә§гҖӮ  
- `dynamic_community_selection=False`пјҡзҰҒз”ЁеҠЁжҖҒзӨҫзҫӨйҖүжӢ©гҖӮ  
- `response_type="Multiple Paragraphs"`пјҡи®ҫзҪ®иҝ”еӣһзҡ„жҹҘиҜўз»“жһңдёәеӨҡж®өиҗҪж јејҸгҖӮ  
- `query=query`пјҡжҹҘиҜўй—®йўҳдёә **"иҜ·её®жҲ‘еҜ№жҜ”дёӢID3е’ҢC4.5еҶізӯ–ж ‘з®—жі•дјҳеҠЈеҠҝгҖӮе№¶з”Ёдёӯж–ҮиҝӣиЎҢеӣһзӯ”гҖӮ"**гҖӮ

**и§ЈжһҗжҹҘиҜўз»“жһң**

`context` еҸҳйҮҸеҢ…еҗ«е…ідәҺжҹҘиҜўиҝҮзЁӢзҡ„иҜҰз»Ҷе…ғж•°жҚ®пјҢеҢ…жӢ¬пјҡ
- жҹҘиҜўиҝҮзЁӢдёӯжЈҖзҙўеҲ°зҡ„ж•°жҚ®дҝЎжҒҜ
- иў«з”ЁдәҺжһ„е»әдёҠдёӢж–Үзҡ„ж–Үжң¬зүҮж®ө
- е…¶д»–е…ғж•°жҚ®  

ж·ұе…ҘеҲҶжһҗ `context` еҜ№иұЎеҸҜд»ҘиҺ·еҸ–жӣҙзІҫз»Ҷзҡ„дҝЎжҒҜпјҢжҜ”еҰӮ**LLM жЁЎеһӢжңҖз»ҲдҪҝз”Ёзҡ„ж–Үжң¬ж•°жҚ®жқҘжәҗ**гҖӮ

#### 3.е°ҒиЈ…еҮҪж•°е®ҢжҲҗGraphRAG Query

````python
async def rag_ML(query: str) -> str:
    """
    иҫ“е…ҘжңәеҷЁеӯҰд№ йўҶеҹҹзӣёе…ій—®йўҳпјҢиҺ·еҫ—й—®йўҳзӯ”жЎҲгҖӮ
    :param query: жңәеҷЁеӯҰд№ йўҶеҹҹзҡ„зӣёе…ій—®йўҳ
    :return: queryй—®йўҳеҜ№еә”зҡ„зӯ”жЎҲ
    """
    PROJECT_DIRECTORY = "/root/autodl-tmp/MCP/mcp-graphrag/graphrag"
    graphrag_config = load_config(Path(PROJECT_DIRECTORY))
    
    # еҠ иҪҪе®һдҪ“
    entities = pd.read_parquet(f"{PROJECT_DIRECTORY}/output/entities.parquet")
    # еҠ иҪҪзӨҫеҢә
    communities = pd.read_parquet(f"{PROJECT_DIRECTORY}/output/communities.parquet")
    # еҠ иҪҪзӨҫеҢәжҠҘе‘Ҡ
    community_reports = pd.read_parquet(
        f"{PROJECT_DIRECTORY}/output/community_reports.parquet"
    )
    # иҝӣиЎҢе…ЁеұҖжҗңзҙў
    response, context = await api.global_search(
        config=graphrag_config,
        entities=entities,
        communities=communities,
        community_reports=community_reports,
        community_level=2,
        dynamic_community_selection=False,
        response_type="Multiple Paragraphs",
        query=query,
    )
    
    return response
````
